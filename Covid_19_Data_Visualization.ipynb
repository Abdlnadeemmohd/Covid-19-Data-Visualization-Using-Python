```python
Covid_cases.drop( ["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational"],inplace= True, axis= 1)
```

```python
Covid_cases.head(10)
```

```python
#Total Cases
Covid_cases['Total Cases'] = (Covid_cases['Confirmed'] + Covid_cases['Cured']+ Covid_cases['Deaths'])
```

```python
#Active Cases
Covid_cases['Active Cases'] = Covid_cases['Confirmed'] - (Covid_cases['Cured']+Covid_cases['Deaths'])
```

```python
#tail to add the column
Covid_cases.tail()
```

```python
Covid_cases.head(2)
```

```python
Statewise = pd.pivot_table(Covid_cases, values= ["Confirmed","Deaths","Cured"], index="State-UnionTerritory",aggfunc= max)
```

```python
Statewise["Recovery Rate"] = Statewise['Cured']*100/Statewise["Confirmed"]
```

```python
Statewise["Mortality Rate"] = Statewise['Deaths']*100/Statewise["Confirmed"]
```

```python
Statewise["Total Cases"] = Statewise['Deaths']+Statewise["Confirmed"]+Statewise["Cured"]
```

```python
Statewise = Statewise.sort_values(by="Confirmed", ascending= False)
```

```python
Statewise.style.background_gradient(cmap="cubehelix")
```

```python
#Grouping

Top_10_States = Covid_cases.groupby(by = 'State-UnionTerritory').max()[['Active Cases','Date']].sort_values(by = ['Active Cases'], ascending= False).reset_index()
fig = plt.figure(figsize=(16,9))
plt.title("Top 10 States with Active cases", size = 25)
ax = sns.barplot(data = Top_10_States.iloc[:10], y = "Active Cases", x = "State-UnionTerritory", linewidth = 1, edgecolor = 'black')
plt.xlabel('States')
plt.ylabel("Active Cases")
plt.show()
```

```python
#Top Deaths

Top_10_Deaths = Covid_cases.groupby(by='State-UnionTerritory').max()[['Deaths','Date']].sort_values(by= ['Deaths'],ascending= False).reset_index()
fig = plt.figure(figsize=(16,9))
plt.title("Top 10 States with Deaths", size = 25)
ax = sns.barplot(data = Top_10_Deaths.iloc[:12], y = "Deaths", x = "State-UnionTerritory", linewidth = 1, edgecolor = 'black')
plt.xlabel('States')
plt.ylabel("Deaths")
plt.show()

```

```python
#remove rows 

Vaccine = Vaccination[Vaccination!='India']
Vaccine
```

```python
Vaccine.info()
```

```python
Vaccine.rename(columns={'Total Individuals Vaccinated' : 'Total'},inplace= True)
Vaccine
```

```python
#most vacinated
max_vac = Vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total',ascending = False)[:5]
max_vac
```

```python
fig = plt.figure(figsize=(8,6))
plt.title("Top 5 Vacinated States", size = 25)
ax = sns.barplot(data = max_vac.iloc[:10], y = max_vac.Total, x = max_vac.index, linewidth = 1, edgecolor = 'black')
plt.xlabel('States')
plt.ylabel("Vaccinated")
plt.show()
```

```python
min_vac = Vaccine.groupby('State')['Total'].sum().to_frame('Total')
min_vac = min_vac.sort_values('Total',ascending=True)[:5]
min_vac
```

```python
fig = plt.figure(figsize=(16,9))
plt.title("Least 5 Vacinated States", size = 25)
ax = sns.barplot(data = min_vac.iloc[:10], y = min_vac.Total, x = min_vac.index, linewidth = 1, edgecolor = 'black')
plt.xlabel('States')
plt.ylabel("Vaccinated")
plt.show()
```
